```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get("https://www.tutiempo.net/delhi.html?datos=por-horas")
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
temp_byhour=[]
for i in soup.find_all("span",class_="h"):
    temp_byhour.append(i.text)

temp_byhour
```

```python
temp=[]
for i in soup.find_all("span",class_="t temp"):
    temp.append(i.text)
temp
```

```python
wind=[]
for i in soup.find_all("span",class_="vv v"):
    wind.append(i.text)
wind
```

```python
print(len(temp_byhour),len(temp),len(wind))
```

```python
import pandas as pd
df= pd.DataFrame({"temp_byhour":temp_byhour,"temp":temp,"wind":wind})
df
```
