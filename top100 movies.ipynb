```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get("https://www.imdb.com/list/ls055592025/")
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
scraped_movies= soup.find("h3",class_="lister-item-header")
scraped_movies
```

```python
scraped_movies.text.replace("\n","")
```

```python
scraped_movies=[]
for i in soup.find_all("h3",class_="lister-item-header"):
    scraped_movies.append(i.text.replace("\n",""))
    
scraped_movies
```

```python
rating=soup.find("span",class_="ipl-rating-star__rating")
rating
```

```python
rating.text
```

```python
rating=[]
for i in soup.find_all('div', class_="ipl-rating-star small"):
    rating.append(i.text.replace('\n',''))


    
rating
```

```python
year=soup.find("span",class_="lister-item-year text-muted unbold")
year
```

```python
year.text
```

```python
year=[]
for i in soup.find_all("span",class_="lister-item-year text-muted unbold"):
    year.append(i.text)
    
year
```

```python
print(len(scraped_movies),len(rating),len(year))
```

```python
import pandas as pd
df=pd.DataFrame({'scraped_movies':scraped_movies,'rating':rating,'year':year})
df
```
